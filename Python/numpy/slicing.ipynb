```python
# Customer id, name
c = np.array([
    [101,"amrendra"],
    [102,"mahendra"],
    [103,"devasena"]
])

# id, purchase amount, date
d = np.array([
    [101,250.50,'2026-08-01'],
    [102,10,'2026-08-24'],
    [103,23,'2026-08-02']
])
```

```python
# combine all the info --> horizontally
comb = np.hstack((c,d))
comb
```

```python
# combines all the info --> vertically
e = np.array([
    [104,'shivagami'],
    [105,'kattappa'],
    [106,'bhallaldeva']
])

comb_e = np.vstack((c,e))
comb_e
```

```python
# Spliting the array into 2
transactions = np.array([
    [101,'joey',10.33,'2018-09-08'],
    [102,'rachel',500.34,'2020-09-03'],
    [103,'monica',900.98,'2020-11-20'],
    [104,'phoebe',178.9,'2020-10-5']
])
```

```python
# horizontal split
code , date = np.hsplit(transactions,[3])
code , date
```

```python
# vertical split
x , y = np.vsplit(transactions, [3])
x , y
```
