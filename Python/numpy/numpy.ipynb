```python
# arrays occupy less memory than list 
python_list = list(range(10))
python_list[:5]
```

```python
import sys
sys.getsizeof(python_list)
```

```python
python_array = np.array(range(10))
python_array[:5]

# Why NumPy Array May Show Bigger Size Initially?
# NumPy arrays have extra internal metadata like:
# datatype info
# shape
# strides
# buffer management
# So for very small arrays (like 10 elements):
# this overhead can make NumPy appear larger.
```

```python
sys.getsizeof(python_array)

# But For Large Data…
# NumPy becomes much more memory efficient.
# Example:
# 1 million integers
# Python list becomes huge because:
# each integer is a separate Python object
# NumPy array stores:
# raw integers compactly
# Huge difference appears.
```

```python
python_array.nbytes
```

```python
l1 = [1,2,3]
l2 = [4,5,6]
l3 = zip(l1,l2)
for i in l3:
    print(i)

# or 
for x,y in zip(l1,l2):
    print(x,y)
```

```python
# arrays are much faster than lists.
import time 

SIZE = 1000
l1 = list(range(SIZE))
l2 = list(range(SIZE))

start_time = time.time()
l3 = [x+y for x,y in zip(l1,l2)]
end_time = time.time()

print(f"Python list took : {end_time - start_time} time to execute.")
```

```python
n1 = np.array(SIZE)
n2 = np.array(SIZE)

start_time = time.time()
n3 = n1 + n2
end_time = time.time()

print(f"Python Numpy array took : {end_time - start_time} time to execute.")
```

```python
# Convinient use with built in fuctions
l1 = [1,2,3]
l2 = [4,5,6]
print(l1 + l2)   # performs concatenation

# for addition 
l3 = []
for x,y in zip(l1,l2):
    l3.append(x+y)
print(l3)
```
