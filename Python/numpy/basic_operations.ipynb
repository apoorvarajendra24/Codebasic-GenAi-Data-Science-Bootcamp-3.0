```python
rev = np.array([[10,12,9],[15,11,13]])
rev.ndim
```

```python
# to convert multi-dim array to 1-dim array
print(rev.ravel())   
print(rev)
# it returns a view (linked to original array)
# changes may affect original copy

a = np.array([1,2,3,4])
x = a.ravel()
x[0] = 20
print(a)
```

```python
print(rev.flatten())
print(rev)

# it returns a copy
# changes do not affect the original array

a = np.array([1,2,3,4])
x = a.flatten()
x[0] = 10
print(a)
```
