```python
import numpy as np
```

```python
# sales data for Quarter 1 (Martix 1)
# Rows represent different products, columns represent different regions

q1 = np.array([
        [200,220,250],       # porduct a
        [150,180,210],       # product b
        [300,330,360]        # product c
    ])

# sales data for Quarter 2 (Matrix 2)
q2 = np.array([
        [209,231,259],       # product a
        [155,192,222],       # product b
        [310,340,375]        # product c
])
```

```python
# total sales in both quarter
q1 + q2
```

```python
# sales relation in quarter
q2 - q1
```

```python
# percentage increase
(q2 -q1)*100 / q1
```

```python
# prices for different products in different region
prices = np.array([
    [10,12,11],
    [8,9,10],
    [15,16,17]
])

q1_revenue = q1 * prices 
q1_revenue
```

```python
q1_discount = q1_revenue * 0.2 # discount of 20% 
q1_discount
```

```python
net_revenue = q1_revenue - q1_discount
net_revenue
```

```python
# dot product (mostly used in ml)

features = np.array([
    [2000,3],    # house 1
    [1800,2]     # house 2
])

weights = np.array([150,50000])

# to predict the price of the house you do 
# 2000*150 + 3*50000

price = np.dot(features,weights)
price
```
