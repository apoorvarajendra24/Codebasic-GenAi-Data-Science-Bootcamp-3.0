```python
# Working with files and do operaions using python code is file handling.

f = open("text.txt",'r')
for line in f:
    print(line)
f.close()
# deallocate the resources that it allocated to open the file.
```

```python
with open('text.txt','r') as f:
    for line in f:
        print(line)

# do not have to use f.close() because when the context of with block ends
# it automatically calls f.close()
```

```python
# read lines in a list.
with open('text.txt','r') as f:
        lines = f.readlines()
        for line in lines:
                print(line)
```

```python
with open("scores.csv",'r') as f:
    lines = f.readlines()
    data = [line.strip().split(',') for line in lines]
    #for line in lines:
    #   a.append(line.strip().split(','))
print(data)
```

```python
players = []
for row in data:
    if row[0] not in players:
        players.append(row[0])
print(players)         
print(type(players[0]))
```

```python
player_info = {}
for player in players:
    player_info[player] = []
    for info in data:
        if info[0] == player:
            player_info[player].append(int(info[2]))
print(player_info)
```

```python
import numpy as np
for key, value in player_info.items():
    player_info[key] = np.mean(value)
print(player_info)
```

```python
# Efficient way
player_info = {}
with open('scores.csv','r') as f:
    for line in f:
        player, _, runs = line.strip().split(',')

        if player not in player_info:
            player_info[player] = []

        player_info[player].append(int(runs))

print(player_info)
```

```python
for player, runs in player_info.items():
    min_score = min(runs)
    max_score = max(runs)
    average_score = sum(runs)/len(runs)
    print(f"{player.upper()} ==> MIN SCORE:{min_score}, MAX_SCORE:{max_score}, AVG:{average_score}")
```
