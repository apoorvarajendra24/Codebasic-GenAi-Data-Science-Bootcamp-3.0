```python
with open("scores.csv",'r') as f:
    lines = f.readlines()
    data = [line.strip().split(',') for line in lines]
    #for line in lines:
    #   a.append(line.strip().split(','))
print(data)
```

```python
players = []
for row in data:
    if row[0] not in players:
        players.append(row[0])
print(players)         
print(type(players[0]))
```

```python
player_info = {}
for player in players:
    player_info[player] = []
    for info in data:
        if info[0] == player:
            player_info[player].append(int(info[2]))
print(player_info)
```

```python
import numpy as np
for key, value in player_info.items():
    player_info[key] = np.mean(value)
print(player_info)
```

```python
# Efficient way
player_info = {}
with open('scores.csv','r') as f:
    for line in f:
        player, _, runs = line.strip().split(',')

        if player not in player_info:
            player_info[player] = []

        player_info[player].append(int(runs))

print(player_info)
```

```python
for player, runs in player_info.items():
    min_score = min(runs)
    max_score = max(runs)
    average_score = sum(runs)/len(runs)
    print(f"{player.upper()} ==> MIN SCORE:{min_score}, MAX_SCORE:{max_score}, AVG:{average_score}")
```

```python
with open("practice.txt","r") as f:
    file = f.readlines()
    found = False
    for lines in file:
        line = lines.strip()
        if "learning" in line:
            found = True
        if "java" in line:
            line = line.replace("java","python")
        print(line)
    if found:
        print("present")
    else:
        print("absent")
```

```python
def change_word_and_find_word(file,word,word_to_find):
    found = False
    with open(file,"r") as f:
        data = f.read()
    new_data = data.replace(word,"python")
    if word_to_find in data:
        found = True
    if found:
        print("Word present")
    else:
        print("Word absent")

    with open(file,"w") as f:
        f.write(new_data)
    print(new_data)
```

```python
def find_word(file,word):
    line_count = 0
    found = False
    with open(file,"r") as f:
        for line in f.readlines():
            line_count += 1
            if word in line:
                found = True
                print(f"Word :{word} found at line :{line_count}")
    if not found:
        return -1
```

```python
value = find_word("practice.txt","learning")
value
```

```python
import random
def print_even_numbers(file):
    numbers = [str(random.randint(1,100)) for _ in range(10)]
    numbers = ",".join(numbers)
    with open(file,"w") as f:
        f.write(numbers)
    
    count_even = 0
    with open(file,"r") as f:
        for num in f.read().split(","):
            if int(num) % 2 == 0:
                count_even += 1

    return f"Number of even numbers in series: {count_even}"
```

```python
value2 = print_even_numbers("numbers.txt")
print(value2)
```
