```python
import pandas as pd
```

```python
df_sales = pd.read_excel('linechart.xlsx')
df_sales
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(df_sales['Quarter'], df_sales['Fridge'], label='Fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], label='Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], label='Washing Mchine')
plt.xlabel("Financial Quarter")
plt.ylabel("Revenue (in mln $)")
plt.title("Product Sales")
plt.legend()
plt.show()
```

```python
# pie chart
total_sales = df_sales[['Fridge','Dishwasher','Washing Machine']].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales,
    labels=total_sales.index, 
    autopct='%1.2f%%', 
    explode=(0.01,0.01,0.1), 
    shadow=True,
    startangle=180)
plt.show()
```

```python
df_sales.plot(kind='bar', x='Quarter')
plt.xticks(rotation=50)
plt.show()
```

```python
df_sales2 = df_sales.set_index('Quarter')
df_sales2
```

```python
df_sales2.plot(kind='bar')
plt.xticks(rotation=50)
plt.show()
```
