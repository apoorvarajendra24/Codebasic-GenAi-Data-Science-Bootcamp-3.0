```python
# ways to remove Outliers
# Using 3 standard deviation to remove outliers.
# samples away by 3 std from the mean.

# calculate mean and std
mean = df_bookings['revenue_generated'].mean()
std = df_bookings['revenue_generated'].std()

mean , std
```

```python
# define the acceptable range
lower = mean - 3 * std
upper = mean + 3 * std

lower , upper

# rows outside these range are treated as outliers.
```

```python
df_bookings[df_bookings.revenue_generated < 0]

# none. therefore, no need to care about lower limit
```

```python
df_bookings[df_bookings.revenue_generated > upper]
```

```python
df_temp = df_bookings[
    (df_bookings['revenue_generated'] >= lower) &
    (df_bookings['revenue_generated'] <= upper)
]
df_temp.revenue_generated.min(), df_temp.revenue_generated.max()
```

```python
# Z-score
# it measures how many standard deviation a value is from the mean

z = (df_bookings['revenue_generated'] - df_bookings['revenue_generated'].mean()) / df_bookings['revenue_generated'].std()
z
```

```python
# standard deviation works best when the data is roughly normally distributed
```

```python
df_z = df_bookings[abs(z) <=3]
df_z.revenue_generated.min() , df_z.revenue_generated.max()
```

```python
# IQR 
# for skewed data, the iqr method is often reliable.

# find Q1 and Q3
Q1 = df_bookings['revenue_generated'].quantile(0.25)
Q3 = df_bookings['revenue_generated'].quantile(0.75)

# calculate iqr
IQR = Q3 - Q1

# calculate lower and upper bounds

# Since revenue cannot be negative, the lower bound being negative simply means:
# There are no low-end outliers according to the IQR rule.
lower_limit = max(0, Q1 - 3 * IQR)
upper_limit = Q3 + 3 * IQR

print(f'IQR : {IQR}')
print(f'LOWER BOUND : {lower_limit}')
print(f'UPPER BOUND : {upper_limit}')

# anything below lower and above upper is outlier
```

```python
outliers = df_bookings[
    #(df_bookings['revenue_generated'] < lower) |
    (df_bookings['revenue_generated'] > upper)
]
outliers['revenue_generated'].value_counts()
```

```python
# filtering out the outliers greater than the upper_limit to the actually dataframe
df_bookings = df_bookings[df_bookings['revenue_generated'] < upper]
df_bookings
```
