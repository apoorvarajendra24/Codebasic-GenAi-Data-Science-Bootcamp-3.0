```python
# Finally Block ----> block where the code inside gets executed, no matter even if exception occurs.
file = None
try :
    file = open("example.txt","r")
    content = file.read()
    print(content)

except FileNotFoundError:
    print("Error!! File Not Found.")

finally:
    if file:
        file.close()
    print("File Closed.")
```

```python
# throwing an exception

balance = 0
def deposit(amount):
    global balance
    if amount <= 0:
        raise ValueError("Amount cannot be deposited. Must be greater than zero.")
    balance += amount
    return balance
val = deposit(1000)
print(val)
val = deposit(400)
print(val)
val = deposit(-10)
print(val)
```

```python
class InsufficientBalance(Exception):
    pass
def withdraw(amount):
    global balance 
    if amount > balance:
        raise InsufficientBalance("Amount not available.")
    balance -= amount
    return balance

val1 = withdraw(10)
print(val1)
val1 = withdraw(13900)
print(val1)
```
