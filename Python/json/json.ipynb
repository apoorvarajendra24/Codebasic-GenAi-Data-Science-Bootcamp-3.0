```python
# Json ---> format used widely for storing and exchanging data(server<-->client).
# earlier XML was used for exchange.
```

```python
# Json string

json_string = """
    {
        "name" : "Amrendra",
        "age" : 30,
        "is_employee" : true,
        "skills" : ["Python","Data Analyst", "Machine Learning"],
        "address" : {
                        "street" : "vinayak road",
                        "city" : "Mahishmati"
                    }   
    }
"""
type(json_string)
```

```python
import json
record = json.loads(json_string)
record

# json.loads()	input --> JSON string    output --> Python object (usually dict/list)
# purpose --> Convert string → Python object
```

```python
type(record)
```

```python
record['skills'][1]
```

```python
data = {
        "name" : "Amrendra",
        "age" : 30,
        "is_employee" : True,
        "skills" : ["Python","Data Analyst", "Machine Learning"],
        "address" : {
                        "street" : "vinayak road",
                        "city" : "Mahishmati"
                    }   
    }
```

```python
data_str = json.dumps(data)
data_str

# json.dumps()  input ---> Python object    output ---> JSON string
# purpose ---> Convert Python object → string
```

```python
type(data_str)
```

```python
# Json object to file
# dict data to a json file.
with open("output.json","w") as f:
    json.dump(data, f, indent = 4)

# json.dump()   input ---> Python object    output --->JSON data written to a file
# purpose ---> Save Python object to a file
```

```python
# to load json file

with open('output.json','r') as f:
    data_json = json.load(f)
print(type(data_json))
print(data_json)

# json.load()   input ---> File containing JSON    output ---> Python object
# purpose --->Read JSON from a file
```

```python
# Malformed json strings

json_string = """
{
        "name" : "krishna",
        "age" : 30
        "is_employeed" : true
}
"""
```

```python
from json import JSONDecodeError


try:
    json.loads(json_string)
except JSONDecodeError as e:
    print(f"Failed To parse JSON : {e}")
```
