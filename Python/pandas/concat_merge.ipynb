```python
df1 = pd.DataFrame({
    'city':['New York','Chicago','Orlando','Baltimore'],
    'temperature':[21,14,35,38]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['New York','Chicago','san diego'],
    'humidity':[65,68,75]
})
df2
```

```python
# by default inner join ---> common values
pd.merge(df1,df2,on='city', how='inner')
```

```python
# left join ---> all values in left df and the common values, but not the values from right df
pd.merge(df1,df2,on='city', how='left')
```

```python
# right join ---> all values in right df and the common values, but not the values from left df
pd.merge(df1,df2,on='city', how='right')
```

```python
# outer join ---> all the values from both the df
pd.merge(df1,df2,on='city', how='outer')
```
