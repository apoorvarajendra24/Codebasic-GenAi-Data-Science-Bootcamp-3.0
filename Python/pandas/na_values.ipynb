```python
# or
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
print(df.dtypes)
df.fillna({
    'temperature': 0.0,
    'windspeed': 0.0,
    'event':"0"
}, inplace=True)
df
```

```python
# better way is to fill with mean/median values. best --> median 
# cause a large value can skew the mean.

df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No event'
}, inplace=True)
df
```

```python
df
```

```python
# forward fill --> replacing nan values with forward value
df.ffill()
```

```python
df.bfill()
```

```python
# Forward fill missing values, but fill at most 1 consecutive NaN 
df.ffill(limit=1)
```

```python
df.bfill(limit=2)
```

```python
# interpolate() --> fill missing (NaN) values by estimating values between existing data points.
# cannot perform with strings
df.temperature.interpolate(method='linear')
```

```python
df.windspeed.interpolate()
```

```python
# dropna ---> drop rows which consists null values
df.dropna()
```

```python
df.dropna(how='any')
```

```python
df.dropna(how='all',inplace=True)
df
```

```python
df.isna()
```

```python
import numpy as np
df.replace("NaN",np.nan, inplace=True)
df
```

```python
# even if one nan remove
df.dropna(thresh=1,inplace = True)
df
```

```python
df1 = pd.read_csv("weather_data_1.csv")
df1
```

```python
# in most of the finance companies, the nan values are represented as -99999.
# to replace that 

df1.replace(-99999,0)
```

```python
# or 
df1.replace([-99999,-88888],np.nan)
```

```python
df1.replace({
    'temperature': -99999,
    'windspeed':[-99999,-88888],
    'event':"no event"
}, np.nan)
```

```python
# also allows us to replace old value to new value
df1.replace({
    -99999:0,
    -88888:0,
    'no event': 'rainy'
})
```
