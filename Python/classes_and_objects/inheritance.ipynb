```python
# Inheritance ---> When one class(child/derived) derives the properties of
# and methods of another class(parent/base).

import datetime
class Cricket_player:
    team_size = 11
    def __init__(self,fname,lname,birth_year,team):
        self.first_name = fname
        self.last_name = lname
        self.birth_year = birth_year
        self.runs = []
        self.team = team
    
    def add_score(self,score):
        return self.runs.append(score)
    
    def average_score(self):
        return sum(self.runs) / len(self.runs)

    def get_age(self):
        now = datetime.datetime.now().year
        return now - self.birth_year
    
    def __str__(self):     # it defines how an object looks like when printed.
        return f"{self.first_name} {self.last_name} is a Cricketer, Who playes for the team {self.team}."
    
    def __lt__(self,other):
        self.avg_score = self.average_score()
        other.avg_score = other.average_score()
        return self.avg_score < other.avg_score
```

```python
class Tennis_player:
    team_size = 11
    def __init__(self,fname,lname,birth_year,team):
        self.first_name = fname
        self.last_name = lname
        self.birth_year = birth_year
        self.aces = []
        self.team = team
    
    def add_aces(self,aces):
        return self.aces.append(aces)
    
    def average_ace_per_match(self):
        return sum(self.aces) / len(self.aces)

    def get_age(self):
        now = datetime.datetime.now().year
        return now - self.birth_year
    
    def __str__(self):     # it defines how an object looks like when printed.
        return f"{self.first_name} {self.last_name} is a Tennis Player, Who playes for the Country {self.team}."
    
    def __lt__(self,other):
        self.avg_ace = self.average_ace_per_match()
        other.avg_score = other.average_ace_per_match()
        return self.avg_score < other.avg_score
```

```python
virat = Cricket_player("Virat","Kohli",1988,"India")
virat.add_score(98)
virat.add_score(101)
virat.add_score(234)
print(virat)
```

```python
roger = Tennis_player("Roger","Federer",1981,"Switzerland")
roger.add_aces(7)
roger.add_aces(10)
roger.add_aces(5)
print(roger)
```

```python
# Single inheritance
class Car():           # <----- Parent class
    color = "black"
    @staticmethod
    def start():
        print("Car started..")
    
    @staticmethod
    def stop():
        print("Car stopped..")
```

```python
class ToyotaCar(Car):    # <------ derived class
    def __init__(self,name):
        self.name = name
```

```python
car1 = ToyotaCar('Fortuner')
car2 = ToyotaCar('Camrey')
print(car1.name)
```

```python
car1.start(), car2.start()
```

```python
print(car1.color)
```

```python
# Multi level Inheritance  ---> base class properties derived by child class and 
# properties of child class also derived by another child class.
# base class ---> derived child class ---> derived child class

# eg, grandfather ---> father ---> Son


class Grandfather():
    def land_property(self):
        print("Grandfather Owns the Land")
    
class Father(Grandfather):
    def car(self):
        print('Father owns the car')

class Child(Father):
    def __init__(self,name):
        self.name = name
    def introduce(self):
        print(f"My name is {self.name}")

child1 = Child('Mahendra')
child1.introduce()
```

```python
# inherited from father
child1.car()
```

```python
# inherited from grandfather
child1.land_property()
```

```python
# father inherited from his father
father1 = Father()
father1.land_property()
```

```python
# Multiple Inheritance ---> a class can be a derived class if it 
# inherits from multiple(2 or more) parent class.

class A:
    def welcome(self):
        print("hello ji")

class B:
    def welcome(self):
        print("bye ji")

class C(B,A):
    pass


# This is also a example of method overriding/method conflict in 
# multile inheritance.ipynb
# that is when both the parent classes hae a method/attr with the same name.

# python must decide which should run and it decides as follows:
# it chooses according to MRO:
# left parent first then right parent
```

```python
c = C()
c.welcome()
```

```python
# Method Overloading
# python does not support true method oveloading directly.

class Math:
    def add(self, a,b):
        print(a+b)
    
    def add(self,a,b,c):
        print(a+b+c)
```

```python
m1 = Math()
m1.add(1,2,3)  # missing 1 argument ie, only the last method exists. i.e, it replaces the first one
```

```python
# method overloading is achieved in python using default agruments or *args

class Calculate:
    def add(self,a,b,c=0):
        return a+b+c
    
    def add(self,*numbers):
        return sum(numbers)
```

```python
cal = Calculate()
print(cal.add(10,20))
print(cal.add(2,4,5))
print(cal.add(1,2,3,4,5))
```

```python
# Super Method  ----> super() is used to access methods of the parnt class.

class Car:
    def __init__(self,type):
        self.type = type

    @staticmethod
    def start():
        print("Car Started..")
    
    @staticmethod
    def stop():
        print("Car Stopped..")
    
class ToyotaCar(Car):
    def __init__(self,name,type):
        super().__init__(type)
        self.name = name
        
    

car1 = ToyotaCar("prius","petrol")
print(car1.name)
print(car1.type)
```

```python
# Class method ---> A class method is bound to the classs and receives the class as an implicit 
# first argument.

# NOTE : static method cant access or modify class state and generally for utlility

class Person:
    name = "anonymous"
    
    #def changeName(self,name):
    #    self.name = name
        # Person.(anything) or self.__class__.(anything) ---> indirect ways 
    
    @classmethod
    def changeName(cls,name):
        cls.name = name


# A class method is used when you want to work with: 1) class attr 2) the class itself
# instead of individual objects.
```

```python
p1 = Person()
p1.changeName('bhalaldeva')
print(p1.name)
print(Person.name)
```

```python
# Example
class Student():
    school_name = "BBA School of BTECH"

    @classmethod
    def change_school(cls,school_name):
        cls.school_name = school_name

s1 = Student()
s2 = Student()

Student.change_school("BCOM Institute of BTECH")

print(s1.school_name)
print(s2.school_name)
print(Student.school_name)
```

```python
# Property 
# We use @property decorator on any method in the class to use the method as a property.

class Student():
    def __init__(self,phy,chem,math):
        self.phy = phy
        self.chem = chem
        self.math = math
    
    #def calcpercentage(self):
    #    self.percentage = str((self.phy + self.chem + self.math) / 3) + "%"
    #    return self.percentage
    
    @property
    def percentage(self):
        return str((self.phy + self.chem + self.math) / 3) + "%"
    
    # When a value of an attribute deoends upon a function, we use it as a property.
```

```python
stu1 = Student(98,76,99)
stu1.percentage
```

```python
stu1.phy = 99
stu1.percentage
```
