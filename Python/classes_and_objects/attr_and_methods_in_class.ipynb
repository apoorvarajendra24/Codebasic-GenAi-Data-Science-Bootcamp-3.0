```python
# Methods are functions that belongs to objects
class Student:
    def __init__(self,name):
        self.name = name        # attributes
        self.marks = []
    
    def welcome(self):          # method
        print(f"Welcome {self.name}")
    
    def add_marks(self,marks):
        return self.marks.append(marks)
    
    def get_marks(self):
        print(f"Marks: {self.marks}")
```

```python
s1 = Student('Bahubali')
s1.add_marks(45)
s1.add_marks(80)
s1.add_marks(98)
s1.welcome()
s1.get_marks()
```

```python
class Clg_student():
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def get_average(self):
        return sum(self.marks) / len(self.marks)
    
    def show(self):
        print(f"{self.name} with marks: {self.marks}")
```

```python
c1 = Clg_student("Amrendra",[100,100,89])
c1.show()
print(c1.get_average())
```

```python
# Static Methods --> methods that dont use the self parameter(work at class level)

class Tree():
    @staticmethod     # Decorator
    def hello():
        print("hello humans...")
    
# Decorators allows us to wrap another function in order to
# extend the behaviour of the wrapped fucntion, without
# permanently modyfying it
```

```python
t1 = Tree()
Tree.hello()
```
