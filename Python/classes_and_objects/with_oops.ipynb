```python
import datetime
class Cricket_player:
    team_size = 11
    def __init__(self,fname,lname,birth_year,team):
        self.first_name = fname
        self.last_name = lname
        self.birth_year = birth_year
        self.runs = []
        self.team = team
    
    def add_score(self,score):
        return self.runs.append(score)
    
    def average_score(self):
        return sum(self.runs) / len(self.runs)

    def get_age(self):
        now = datetime.datetime.now().year
        return now - self.birth_year
    
    def __str__(self):     # it defines how an object looks like when printed.
        return f"{self.first_name} {self.last_name} is a Cricketer, Who playes for the team {self.team}."
    
    def __lt__(self,other):
        self.avg_score = self.average_score()
        other.avg_score = other.average_score()
        return self.avg_score < other.avg_score

virat = Cricket_player('virat','kohli',1988,"India")
virat.add_score(40)
virat.add_score(239)
virat.add_score(50)
print(virat.average_score())
print(virat.first_name)
print(virat.last_name)
print(virat.runs)
print(virat.get_age())
```

```python
david = Cricket_player('david','warner',1985,'Australia')
david.add_score(234)
david.add_score(100)
david.add_score(32)
print(david.average_score())
print(david.first_name)
print(david.last_name)
print(david.get_age())
print(david)
```
