```python
virat = {
    'first_name': "Virat",
    'second_name': "Kohli",
    'runs': [],
    'birth_year':1988
}
print(virat)
```

```python
virat['runs'].append(90)
virat['runs'].append(89)
virat['runs'].append(134)
```

```python
virat
```

```python
def average_score(player):
    return sum(player['runs']) / len(player['runs'])
```

```python
import datetime
def current_age(player):
    current_year = datetime.datetime.now().year
    return current_year - player['birth_year']
```

```python
print(f"VIRAT AVERAGE SCORE ---> {average_score(virat)}")
print(f"VIRATE AGE ---> {current_age(virat)}")
```

```python
david = {
    'first_name': "David",
    'second_name': "Warner",
    'runs': [84,300,12,90],
    'birth_year':1984
}

print(average_score(david))
print(current_age(david))
```
