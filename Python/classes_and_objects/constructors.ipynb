```python
# Constructor runs automatically when an object is created.
class Student:
    # Default constructor ---> A constructor that takes no extra values from user.
    def __init__(self):
        self.name = "kattappaa"  # Values are fixed/default
        print("default constructor initiated..")
    
```

```python
s1 = Student()
print(s1.name)
```

```python
class Student1:
    college_name = "mahishmati college of engineering"              # Class attribute --> Common to all the objects. (student can be in same college)
    # Parameterized constructor ---> A constructor that takes values from user during object creation.
    def __init__(self, name, marks):
        self.name = name                   # Setting up data for object
        self.marks = marks                 # Instance attribute  --> Specific to object. (names can be different of each student) You use self for instance variable
        print("Parameterized constructor initiated..")
```

```python
s2 = Student1('bahubali',100)
print(s2.name, s2.marks)
print(s2.college_name)
```

```python
# Both default and parameterized constructors cannot be in same class.
# But we can achieve that too

class Car:
    name = "annonymous"                         # obj attr precedence is greater than cls attr, therefore bmw printed.
    def __init__(self, name="unknown"):
        self.name = name

c1 = Car()
c2 = Car('BMW')
print(c1.name)
print(c2.name)
```
