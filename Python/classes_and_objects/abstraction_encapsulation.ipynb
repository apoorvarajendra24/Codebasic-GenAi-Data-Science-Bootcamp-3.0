```python
class Car:
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = False
    
    def start(self):
        self.clutch = True
        self.acc = True
        print("Car Started")
    
car1 = Car()
car1.start()
```

```python
# Encapsulation ---> Wrapping data and functions into single unit(object)
# class with methods and attributes itself is example of encapsulation.
```

```python
class Account:
    def __init__(self,account_no,balance):
        self.account_no = account_no
        self.balance = balance
    
    def debit(self,debit_amount):
        self.balance = self.balance - debit_amount
        print(f"Rupees {debit_amount} debited. Available balance is {self.balance}") 

    def credit(self,credit_amount):
        self.balance = self.balance + credit_amount
        print(f"Rupees {credit_amount} credited. Available balance is {self.balance}")
    
    def show_balance(self):
        return f"Available balance : {self.balance}"
```

```python
user1 = Account(1111000099458,3000)
user1.show_balance()
```

```python
user1.debit(400)
```

```python
user1.credit(10000)
```

```python
user1.show_balance()
```

```python
user2 = Account(1111000099460,10000)
user2.credit(1459)
```

```python
user2.show_balance()
```

```python
# Private(like) attributes and methods

# Private attributes and methods are ment to be used only within 
# the class and are not accessible from outside the class.

 
class Login:
    def __init__(self,userid,pwd):
        self.userid = userid
        self.__pwd = pwd           # private attribute (__pwd)
    
    def __reset_pwd(self):
        print(self.__pwd)
```

```python
u1 = Login('Kattappaa_taklaa','iloveyoou_rajmata')
print(u1.userid)
print(u1.pwd)
u1.reset_pwd()
```

```python
class Person:

    def __init__(self, name):
        self.__name = name
    
    def __hello(self):
        print(f"Hello {self.__name}")
    
    def welcome(self):
        self.__hello()
```

```python
p1 = Person('bijjaldeva')
p1.welcome()
```
