```python
# Complex number class
class ComplexNum:
    def __init__(self,real,img):
        self.real = real
        self.img = img
    
    def showNum(self):
        print(self.real,"i +",self.img,"j")
    
    def __add__(self,num2):
        newReal = self.real + num2.real
        newImg = self.img + num2.img
        return ComplexNum(newReal, newImg)
    
    def __sub__(self,num2):
        newReal = self.real - num2.real
        newImg = self.img - num2.img
        return ComplexNum(newReal, newImg)
```

```python
n1 = ComplexNum(1,3)
n1.showNum()
```

```python
n2 = ComplexNum(4,6)
n2.showNum()
```

```python
# How to add these 2 complex numbers
n3 = n1.add(n2)
n3.showNum()
```

```python
print(n1 + n2)  # no logic to add complex numbers unlike in integer and string
# therefore use dunder functions
```

```python
n3 = n1 + n2
n3.showNum()
```

```python
n4 = n1 - n3
n4.showNum()
```

```python
import math
class Circle:
    def __init__(self,radius):
        self.radius = radius
    
    def area(self):
        return ((math.pi)*(self.radius**2))
    
    def perimeter(self):
        return (2 * math.pi * self.radius)
```

```python
c1 = Circle(10)
print(c1.area())
print(c1.perimeter())
```

```python
class Employee:
    def __init__(self, role, dept, sal):
        self.role = role
        self.dept = dept
        self.sal = sal
    
    def show_details(self):
        print(f"Role : {self.role}")
        print(f"Department : {self.dept}")
        print(f"Salary : {self.sal}")

class Engineer(Employee):
    def __init__(self,name,age):
        super().__init__('God','Feku',40000000)
        self.name = name
        self.age = age
```

```python
emp1 = Employee("AI Engineer","AI/ML Department",800000)
emp1.show_details()
```

```python
eng1 = Engineer('Narendra',75)
eng1.show_details()
```
