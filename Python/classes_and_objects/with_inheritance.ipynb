```python
import datetime
```

```python
# Parent Class ---> can have child class/s
class Player: 
    def __init__(self,fname,lname,birth_year):
        self.first_name = fname
        self.last_name = lname
        self.birth_year = birth_year
    
    def get_age(self):
        now = datetime.datetime.now().year
        return now - self.birth_year
```

```python
# Child class derived from parent class
class TennisPlayer(Player):
    def __init__(self,fname,lname,birth_year):
        super().__init__(fname,lname,birth_year)
        self.aces = []
    
    def add_aces(self,aces):
        return self.aces.append(aces)

    def get_average(self):
        return sum(self.aces)/len(self.aces)

```

```python
roger = TennisPlayer('Roger','Federer',1981)
roger.add_aces(7)
roger.add_aces(10)
print(roger.first_name)
print(roger.last_name)
print(roger.birth_year)
print(roger.get_age())
```

```python
# Child class derived from parent class
class CricketPlayer(Player):
    def __init__(self,fname,lname,birth_year,team):
        super().__init__(fname,lname,birth_year)
        self.team = team
        self.scores = []
    
    def add_score(self,score):
        return self.scores.append(score)

    def get_average(self):
        return sum(self.scores)/len(self.scores)
```

```python
virat = CricketPlayer("Virat","Kohli",1988,"India")
virat.add_score(66)
virat.add_score(0)
virat.add_score(22)
print(virat.get_age())
```
