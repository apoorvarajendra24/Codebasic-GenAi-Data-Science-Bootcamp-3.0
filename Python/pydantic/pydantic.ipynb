```python
class User:
    def __init__(self, user_id, name, age):
        self.id = user_id
        self.name = name
        self.age = age
```

```python
bat = User(10, "Batman", "30 years")
bat.age
```

```python
from pydantic import BaseModel
from typing import Optional
class User(BaseModel):
    # no need to have init method in this, pydantic will internally create init method
    id : int
    name : str
    age : int
    address : Optional[str] = None

    # this is type hint
```

```python
batwom = User(id=11,name="Batwomen",age=30)
batwom
```

```python
lan = User(id=12, name="lantern", age=29, address="124 Vadapav street")
lan
```
