```python
# logging ---> It is a journal for program. keeps information about what the program is doing while it runs.
import logging

# Instead of printing messages on the screen with print(), logging stores useful information such as:
# When the program started
# Which functions were executed
# Important events
# Warnings
# Errors and exceptions
```

```python
# setting up minimum severity level
# The configured logging level acts like a filter: show messages whose numeric value is greater than or equal to the configured level.
logging.basicConfig(
    level = logging.INFO,
    filename= "app.log",       # saves the logs in a file
    force= True,
    format = "%(asctime)s - %(levelname)s - %(message)s"
)

# detailed info for developer    ---> numeric value - 10
logging.debug("this is a debug message")
# normal program events     ---> numeric value - 20
logging.info("this is a info message")
# just a warning nothing serious but the program continues     ---> numeric value - 30
logging.warning("this is a warning message")
# an operation/function fails      ---> numeric value - 40
logging.error("this is an error message")
# severe error, application completely shutsdown       ---> numeric value - 50
logging.critical("this is a critical message")
```
