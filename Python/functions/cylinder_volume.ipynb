```python
#volume of cylinder --> pi * (radius)^2 * h

r = 10
h = 7
print(3.14*(r**2)*h)
```

```python
def cylinder_volume(radius,height):     # you can fix value for parameters if you dont want to pass it as(height = 6)
    print("radius: ",radius)
    print("height: ",height)
    volume = 3.14 * (radius**2) * height
    return volume
```

```python
radius = int(input("Enter radius: "))
height = int(input("Enter height: "))
print(f"Volume of cylinder for radius {radius} and height {height} is: {cylinder_volume(radius,height)}")
```
