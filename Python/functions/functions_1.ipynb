```python
expenses_a = [30,45,70,90]
expenses_b = [40,23,10,85]
```

```python
total_expenses_a = 0
for expense in expenses_a:
    total_expenses_a +=expense
print("total expense for a: ",total_expenses_a)

total_expenses_b = 0
for expense in expenses_b:
    total_expenses_b +=expense
print("total expense for b: ",total_expenses_b)
```

```python
# Function --> code which can be reused multiple times.
def total_expenses(expenses):
    total = 0
    for expense in expenses:
        total += expense
    return total
```

```python
print("Total expense for a:",total_expenses(expenses_a))
print("Total expense for b:",total_expenses(expenses_b))
```
