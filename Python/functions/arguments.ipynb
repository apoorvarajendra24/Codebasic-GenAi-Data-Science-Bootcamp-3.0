```python
def sum_all(a,b,c,d):
    sum = 0
    sum = a + b + c + d
    return sum
a = 1
b = 2
c = 3
d = 4
sum_all(a,b,c,d)
```

```python
def total_of(*args):        # positional arguments
    total = 0
    print(type(args))       # converts arguments into a tuple
    for num in args:
        total += num
    return total
```

```python
total_of(10,20,30,40,12)
```

```python
def company_info(**kwargs):        # keyword arguments
    print(type(kwargs))            # converts arguments into dictionary
    for key in kwargs:
        print(key, kwargs[key])

company_info(ticker="APPL", ceo="Tim Cook", revenue="200 billion", pe=20)
```

```python
def company_info(**kwargs):
    if 'ticker' in kwargs:
        print("Ticker:",kwargs['ticker'])
    if 'ceo' in kwargs:
        print("Ceo:",kwargs['ceo'])
    if 'revenue' in kwargs:
        print("revenue:",kwargs['revenue'])
company_info(ticker="APPL", ceo="Tim Cook", revenue="200 billion", pe=20)
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
print_info(name='rahul',age=28)
```

```python
# lambda function  ---> quick way to write function in one line
# syntax --->   lambda args: expression 

def add(a,b):
    return a+b
a = 10
b = 20

addition = lambda a,b: a+b
print(addition(10,20))
```
