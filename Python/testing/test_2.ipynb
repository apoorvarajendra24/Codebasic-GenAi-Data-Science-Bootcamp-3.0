```python
import pytest
```

```python
def multiply(a,b):
    return a * b

def divide(a,b):
    if b == 0:
        raise ValueError("cannot divide by zero")
    return a/b
```

```python
def test_multiply():
    assert multiply(3,4) == 12
    assert multiply(0,5) == 0
    assert multiply(-1,5) == -5
```

```python
def test_divide():
    assert divide(10,2) == 5
    with pytest.raises(ValueError):
        assert divide(1,0)
```

```python
test_multiply()
```

```python
test_divide()
```
