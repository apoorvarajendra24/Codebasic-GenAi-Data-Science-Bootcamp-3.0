```python
# The assert keyword is used to check whether a condition is True during program execution.
# If the condition is True, the program continues normally.
# If the condition is False, Python raises an AssertionError.

x = 10
try:
    assert x > 0
    print("Program continues")
    assert x < 1 , "x should be less than 1"
except AssertionError as e:
    print(f"Program execution stopped : {e}")
```
