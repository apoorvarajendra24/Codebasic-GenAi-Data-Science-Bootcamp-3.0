```python
def simple():
    data = [
        {'category': 'Food',          'total': 1200.0},
        {'category': 'Shopping',      'total': 2470.0},
        {'category': 'Entertainment', 'total': 225.0},
    ]
    return data
```

```python
def simple_1():
    data = simple()

    total = sum([i['total'] for i in data if not None])
    print(f"total is : {total}")
    return data
```

```python
simple_1()
```
