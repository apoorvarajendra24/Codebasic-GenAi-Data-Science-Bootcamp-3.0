```python
# list --> is a data structure which allows us to store sequential data.
# list is ordered, meaning each item has a fixed position unless explicitly changed.
# list is homogeneous ie can have different data in it.
# list is mutable, allow addition, removal, change within same list.
items = ['bread','pasta','fruits','milk','veggies']
items
```
