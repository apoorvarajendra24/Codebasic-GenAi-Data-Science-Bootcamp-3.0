```python
def find_pe_and_pb(price, eps, book_value):
    pe = price/eps
    pb = price/book_value
    return pe, pb    # ----> it will return a tuple
```

```python
value_pe_ratio, value_pb_ratio = find_pe_and_pb(100,2,4)   #----> unpacking
print(value_pe_ratio)
print(value_pb_ratio)
```
