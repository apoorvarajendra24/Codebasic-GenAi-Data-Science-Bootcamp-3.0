1. Set ---> A set in python is a collection of unique elements. ie, you cannit add duplicate values unlike list. If duplicate added , they are removed automatically.
2. Sets are unordered. 
3. Sets are mutable ---> addition and removal is allowed

```python
myset = set([1,2,4,3,4])
myset
```

```python
# One use case is, If we have a list with duplicate elements, we can use set to remove them and make it a list again,
list(myset)
```

```python
# not subscriptable means they does not have any order.
myset[0]

# which also means a Set Does Not Support Indexes.
```

```python
myset= {1,2,3,4,4,5,6,7}
myset
```

```python
menu = set(['Dosa','Idli','Biryani','Pizza'])
menu
```

```python
menu.add('Burger')
menu
```

```python
menu.add('AlooBonda')
menu
```

Sets are unordered collections. The order in which elements appear when a set is printed is determined by Python's internal hashing mechanism and is not guaranteed to be alphabetical, insertion order, or any fixed order. If a specific order is required, the set must be converted to a sorted list using sorted(). Which also means sets are implemented using hash tables. ie, each element is placed according to its hash value, and not according to given above.

```python
menu.remove('AlooBonda')
```

```python
transactions = [
    {'customer_name': 'Mira', 'amount': 50.33, 'date': "2026-03-01"},
    {'customer_name': 'Mira', 'amount': 12.33, 'date': "2026-08-05"},
    {'customer_name': 'Mohan', 'amount': 21, 'date': "2026-08-03"},
    {'customer_name': 'Mohan', 'amount': 36.9, 'date': "2026-09-10"},
    {'customer_name': 'Lucy', 'amount': 77.9, 'date': "2026-09-10"}
]
```

```python
customer = set()
for trans in transactions:
    customer.add(trans['customer_name'])
print(customer)
```

```python
customer_list = list(customer)
type(customer_list)
```

```python
# Set operations
a = {'Mira','Mohan','Lucy','Amar'}
b = {'Chen','Mohan','Radha','Maira'}
```

```python
# intersection ---> common elements among the two sets
a & b
```

```python
a.intersection(b)
```

```python
# union ---> all unique elements
a | b
```

```python
a.union(b)
```

```python
# difference ---> all in set a but not in set b. common elements are excluded from the result too.
a - b
```

```python
a.difference(b)
```
