```python
expenses = [12000,1300,1500,1700]

total_expenses = expenses[0] + expenses[1] + expenses[2] + expenses[3]
total_expenses
```

```python
total_expenses = 0
for expense in expenses:
    total_expenses+= expense
total_expenses
```

```python
total_expenses = 0
for i, expense in enumerate(expenses):
    print(f"Month: {i+1} ---> Expenses: {expense}")
    total_expenses += expense
print(total_expenses) 
```

```python
# Control Statements
# Break  ----> Stops loop completely
l1 = [i for i in range(5)]
for i in l1:
    if i == 3:
        break
    print(i)
```

```python
# Continue  -----> skips current iteration
for i in l1:
    if i == 2:
        continue
    print(i) 
```

```python
# Pass  ----> eat five star, Does nothing(placeholder)
for i in range(5):
    pass
```

```python
for i in range(3):
    print(i)
else:
    print("completed without break")
```

```python
# Nested for loop
for i in range(3):
    for j in range(3):
        print(i,j,)

```
