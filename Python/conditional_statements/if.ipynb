```python
# odd even condition
n = int(input("Enter a number :"))
if n % 2 == 0:
    print(f"{n} is Even Number")
else:
    print(f"{n} is Odd Number")
```

```python
message = "Number is Even" if n % 2 == 0 else "Number is Odd"
message
```

```python
indian = ["samosa","aloo bonda","dosa"]
chinese = ['egg role','pot sticker', 'fried rice']
italian = ['pizza','pasta','risotto']

print("--------Menu---------")
print("\nIndian:")
for item in indian:
    print("-",item)

print("\nChinese:")
for item in chinese:
    print("-",item)

print("\nItalian:")
for item in italian:
    print("-",item)
```

```python
dish = input("Enter what you want to eat: ")
menu = indian + chinese + italian
if dish in menu:
    if dish in indian:
        print(f"{dish} is an Indian dish!")
    elif dish in chinese:
        print(f"{dish} is a Chinese dish!")
    elif dish in italian:
        print(f"{dish} is an Italian dish!")
else:
    print("Dish not in Menu!")
```
