Iterators allow data to be processed one item at a time. When reading a large file, the file object acts as an iterator, so a for loop reads lines sequentially instead of loading the entire file into memory, making it memory-efficient.

```python
# Iterators ---> they are needed to let python process items one at a time instead of loading everything into memory at once.

prices = []

with open('pricing.txt','r') as file:
    for line in file:
        ticker, country, price, date = line.split('|')
        prices.append({
            'ticker': ticker,
            'country' : country,
            'price' : price,
            'date' : date.strip()
        })
prices[0]['date']
```

```python
# The memory problem would happen if explicitly loaded everything ---> read() or readlines()
with open('pricing.txt','r') as file:
    lines = file.readlines()

# For a very large file, this can consume a lot of RAM.
```

```python
with open('pricing.txt','r') as file:
    for line in file:
        print(line)

# Python does not load the entire file into memory when you open it.
# open() creates a file object.
# The file object is an iterator.
# The for loop repeatedly calls next() on that iterator.
# Each call reads the next line (or a small buffered chunk behind the scenes) as needed.

# only a small portion of the file is kept in memory at any given time.
```

```python
class RemoteControl:
    def __init__(self):
        self.channels = ['NDTV','CNN','BBC','HBO','CNX']
        self.index = -1

    def __next__(self):
        self.index += 1
        if self.index == len(self.channels):
            raise StopIteration
        return self.channels[self.index]
    
    def __iter__(self):
        return self
```

```python
rc = RemoteControl()
for channel in rc:
    print(channel)
```

```python
rc1 = RemoteControl()
next(rc1)
next(rc1)
next(rc1)
next(rc1)
next(rc1)
next(rc1)
```

A generator is a special type of function that produces values one at a time instead of returning all values at once.

It uses the keyword yield instead of return.

```python
# normal function

def number():
    return [1,2,3,4]

print(number())

# the entire list is created and returned at once.
```

```python
# generator 
def numbers():
    yield 1
    yield 2
    yield 3
    yield 4
```

```python
g = numbers()
print(g)

# a generator object is created, but no values are produced yet.
```

```python
print(next(g))
print(next(g))
print(next(g))
print(next(g))
# Each next() resumes execution until the next yield
print(next(g))

# last gives us StopIteration means iteration ended.
```

```python
nums = [i for i in range(10000)]

# this tries to store all 10K numbers in memory
```

```python
def nums():
    for i in range(10000):
        yield i
# only one number is generated when needed.
# much more memory efficient

g = nums()  # g is a generator now
type(g)
```

```python
print(next(g))
print(next(g))
```

```python
for x in nums():
    if x == 10:
        break
    print(x)
```

```python
# a short way to get values which we used a class to get above.
def remote_control():
    yield 'NDTV'
    yield 'CNN'
    yield 'ESPN'

g = remote_control()
for channel in g:
    print(channel)
```

```python
def get_comp_data():
    data = []
    for i in range(3):
        with open(f'stock_{i+1}.txt','r') as f:
            content = f.read()
            data.append(content)
    return data

get_comp_data() 

# for a huge amount of files this code might create an issue for insufficient storage for processing.
```

```python
def get_comp_data():
    for i in range(3):
        with open(f'stock_{i+1}.txt','r') as f:
            content = f.read()
            yield content
```

```python
for i in get_comp_data():
    print(i)
```

```python
# or 
g = get_comp_data()
print(next(g))
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a+b
```

```python
g = fibonacci()
```

```python
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```
