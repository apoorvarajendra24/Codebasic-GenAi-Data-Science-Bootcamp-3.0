```python
import time
```

```python
def get_square(number):
    start = time.time()
    result = []
    for n in number:
        result.append(n*n)
    end = time.time()
    print(f'get_square took {(end - start)*1000} mili seconds')
    return result
```

```python
_ = get_square(range(10))
```

```python
def get_cube(number):
    start = time.time()
    result = []
    for n in number:
        result.append(n*n*n)
    end = time.time()
    print(f'get_cube took {(end - start)*1000} mili seconds')
    return result
```

```python
_ = get_cube(range(10))
```

```python
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__}  took {(end - start)*1000} mili second")
        return result
    return wrapper
    
```

```python
@timer
def get_square(numbers):
    result = []
    for n in numbers:
        result.append(n*n)
    return result    
```

```python
@timer
def get_cube(numbers):
    result = []
    for n in numbers:
        result.append(n*n*n)
    return result
```

```python
_ = get_square(range(10))
```
